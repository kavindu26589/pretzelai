```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range for continuous signal
t_continuous = np.linspace(0, 8 * 2 * np.pi, 10000)  # 8 cycles

# Define the signals
signals = [
    {"A": 44 * np.pi, "omega": 0},
    {"A": 32 * np.pi, "omega": np.pi / 4},
    {"A": 2 * np.pi, "omega": -np.pi / 4},
    {"A": np.pi / 2, "omega": 1}
]

# Sampling frequency
sampling_factor = 7
sampling_frequencies = [abs(signal["omega"] / (2 * np.pi)) * sampling_factor for signal in signals]

# Calculate sampling intervals
sampling_intervals = [1 / f if f != 0 else 1 / sampling_factor for f in sampling_frequencies]
# Sample the signals
sampled_signals = [signal["A"] * np.cos(signal["omega"] * t_continuous) for signal in signals]

# Sample the time axis
t_sampled = [np.arange(0, len(signal) * interval, interval) for signal, interval in zip(sampled_signals, sampling_intervals)]

# Plot the sampled signals
colors = ['b', 'g', 'r', 'c']

plt.figure(figsize=(12, 8))
for i, (signal, t_sample, color) in enumerate(zip(sampled_signals, t_sampled, colors)):
    plt.stem(t_sample, signal, linefmt=color, markerfmt='o', label=f"A = {signals[i]['A']}, omega = {signals[i]['omega']}")

plt.title('Sampled (Discrete) Signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range with a similar resolution
t = np.arange(-10, 10, 0.01)

# Define the period T
T = 2  # Time period (2 time units)

# Create a single bipolar pulse
single_bipolar_pulse = np.zeros_like(t)
single_bipolar_pulse[np.abs(t % T) < 0.005] = 1
single_bipolar_pulse[np.abs(t % T - T) < 0.005] = -1

# Create a data vector with ones and -1s at positions corresponding to the modified bipolar pulse train
data_vector_modified = np.zeros_like(t)
data_vector_modified[np.abs(t % T) < 0.005] = 1
data_vector_modified[np.abs(t % T + 0.5*T) < 0.005] = -1

# Convolve the single bipolar pulse with the modified data vector to obtain the modified bipolar pulse train
bipolar_pulse_train_modified = np.convolve(data_vector_modified, single_bipolar_pulse, 'same')

# Plot the modified bipolar pulse train
plt.figure(figsize=(12, 6))
plt.stem(t, bipolar_pulse_train_modified, linefmt='r-', markerfmt='ro', basefmt=" ", use_line_collection=True)
plt.title('Modified Bipolar Pulse Train Obtained by Convolution')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range with a similar resolution
t = np.arange(-10, 10, 0.01)

# Define the signal x(t) (for demonstration purposes, let's use a simple sinusoidal signal)
x_t = np.sin(2 * np.pi * t)

# Define the shifted signal x(t-2)
x_t_shifted = np.sin(2 * np.pi * (t - 2))

# Compute the resulting signal x(t) + x(t-2)
result_signal = x_t + x_t_shifted

# Plot the resulting signal x(t) + x(t-2)
plt.subplot(3, 1, 3)
plt.plot(t, result_signal, label='x(t) + x(t-2)', color='r')
plt.title('Resulting Signal x(t) + x(t-2)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range with a similar resolution
t = np.arange(-10, 10, 0.01)

# Define the signal x(t) (for demonstration purposes, let's use a simple sinusoidal signal)
x_t = np.sin(2 * np.pi * t)

# Compute the integral of the signal x(t)
integral_x_t = np.cumsum(x_t) * (t[1] - t[0])  # Cumulative sum times the time step

# Plot the integral of the signal x(t)
plt.subplot(2, 1, 2)
plt.plot(t, integral_x_t, label='Integral of x(t)', color='r')
plt.title('Integral of Signal x(t)')
plt.xlabel('Time (t)')
plt.ylabel('Cumulative Amplitude')
plt.grid(True)
plt.legend()

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
